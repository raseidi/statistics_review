### Skewness
- Positive skew (highly-skewed) normally have larger means than medians
- Skewness value greater than zero means that there is more weight in the right tail of the distribution.
- It is sometimes referred to as the third moment about the mean
- When a tail (right or left) extends further than the opposite tail.
- Positive skew (long tail to positive direction) and negative skew (long tail to negative direction)

\begin{aligned}
\begin{gather*}
\sum{\frac{(X - \mu)^3}{\sigma ^ 3}}
\end{gather*}
\end{aligned}

### Kurtosis
- kurtosis is similar to the definition of skew. The value “3” is subtracted to define 3 “no kurtosis” as the kurtosis of a normal distribution. Otherwise, a normal distribution would have a kurtosis of 3.

\begin{aligned}
\begin{gather*}
\sum{\frac{(X - \mu)^4}{\sigma ^ 4}} - 3
\end{gather*}
\end{aligned}


# Discrete Random Variables

## Binomial (_Bernoulli_) distribution
- The distribution of the number *m* of observations of $x=1$, given that the data set has size $N$.
- The probability mass funciton (PMF) of the number of successes (e.g. $x=1$) in a sequence of _n_ trials, where
    - Trials aren indenpendent;
    - Each trial has only two possible outcomes (usually named as success ($x=1$) or failure ($x=0$));
    - Given by the formula:

\begin{aligned}
\begin{gather*}
f(m|N,\mu) = \frac{N!}{m!(N-m)!} \mu^m (1-\mu)^{N-m}
\end{gather*}
\end{aligned}

where you can read it as the number of ways of having *m* occurrencies out of a total of _N_ identical objects, times the probability distribution over a random variable *m*. 

Consider *x* a binary random variable, describing the outome of a flipping coin, with $x=1$ is head, and $x=0$ represents tails. The probability of $x=1$, denoted by the paramter \mu, is given by $p(x=1|\mu) = \mu$, where $0 \leq \mu \leq 1$. Thus, we conclude that $p(x=0|\mu) = 1-\mu$. The probability distribution over $x$ can be written as

\begin{aligned}
\begin{gather*}
f(m|\mu) = \mu^m (1-\mu)^{N-m},
\end{gather*}
\end{aligned}

and the number of possible outcomes for $x$ is two (head or tails) is the number of combinations of *n* elements taking *m* by *m*:


\begin{aligned}
\begin{gather*}
{n\choose k} = \frac{N!}{m!(N-m)!}
\end{gather*}
\end{aligned}

See an example below. Consider a sample of 10 products, where the chance of the product presents a defect is 0.25 ($N=10$ and $\mu=0.25$). The plot shows the distribution probability of having *m* defect products. Thus, we have 25% of change of having 2 defect products.

```python
import matplotlib.pyplot as plt
from math import factorial as fact

N = 10
mu = 0.25
distribution = []
for m in range(N+1):
    f = (fact(N) / (fact(m)*fact(N-m))) * ((mu ** m) * ((1-mu) ** (N-m)))
    distribution.append(f)

distribution = list(map(abs, distribution))
plt.xlabel('Number of defect products (m)')
plt.ylabel('Probability')
plt.bar(x=range(N+1), height=distribution)
```
